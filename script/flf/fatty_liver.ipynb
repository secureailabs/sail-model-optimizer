```python
import pygraphviz as pgv
import matplotlib.pyplot as plt

# Create a new graph
graph = pgv.AGraph(directed=True)

# Add classes and their relationships to the graph
graph.add_node('Person', shape='box')
graph.add_node('Student', shape='box')
graph.add_node('Teacher', shape='box')
graph.add_node('Course', shape='box')

graph.add_edge('Student', 'Person')
graph.add_edge('Teacher', 'Person')
graph.add_edge('Course', 'Teacher')

# Set graph attributes
graph.node_attr['fontname'] = 'Arial'
graph.node_attr['fontsize'] = 12
# Create an interactive viewer
viewer = pgv.graphviz_draw(graph, prog='dot', format='svg')

# Display the viewer in matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(plt.imread(viewer))
plt.axis('off')
plt.show()
```

```python
import pygraphviz as pgv
import matplotlib.pyplot as plt

# Create a new graph
graph = pgv.AGraph(directed=True)

# Add classes and their relationships to the graph
graph.add_node('Person', shape='box')
graph.add_node('Student', shape='box')
graph.add_node('Teacher', shape='box')
graph.add_node('Course', shape='box')

graph.add_edge('Student', 'Person')
graph.add_edge('Teacher', 'Person')
graph.add_edge('Course', 'Teacher')

# Set graph attributes
graph.node_attr['fontname'] = 'Arial'
graph.edge_attr['fontname'] = 'Arial'

# Create an interactive viewer
viewer = pgv.graphviz_draw(graph, prog='dot', format='svg')

# Display the viewer in matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(plt.imread(viewer))
plt.axis('off')
plt.show()
```
